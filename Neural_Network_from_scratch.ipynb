```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
print('versionof numpy',np.__version__)

#LAIIIT: Load data, archtechture, intialize weight, intiate fp, bp, train the model
```

```python
#creating input array

X=np.array([[1,0,1,0],[1,0,1,1],[0,1,0,1]])
print('input')
print(X)
print('\n Shape of Input',X.shape)
```

```python
X=X.T
print('\n Input in matrix form')
print(X)#4 features and 3 observations
```

```python
y=np.array([[1],[1],[0]])
print('\n Actual output \n', y, "\n Shape",y.shape)
#output in matrix form
print("\n", y.T,"\n Shape" ,(y.T).shape)
Y=y.T
```

```python
#Architechture of the model
input_layer_neurons=X.shape[0] #no. of features in data set
hidden_layer_neurons= 3 #no hard n fast rule to choose hidden layer neuron
output_layer_neuron= 1 #No. of ouput for one set of observation
```

```python
#zz#Implementing forward propagation
```

```python
#Intializing weight for all three layers:
#w_ih:Shape of w_ih: no. of neurons in input lyer(features) x number of neurons at hidden layer

w_ih=np.random.uniform(size=(input_layer_neurons,hidden_layer_neurons))
print(w_ih)

#w_ho:Shape of w_ih: number of neurons at hidden layer x no.of neurons at output layer

w_ho=np.random.uniform(size=(hidden_layer_neurons,output_layer_neuron))
print(w_ho)
```

```python
#Implementing forward propagation

#Definingsigmoid ativation function:
def sigmoid (x):
  return 1/(1 + np.exp(-x)) #np.exp -->> e
```

```python
#z1
hidden_layer_input= np.dot(w_ih.T,X) #(3 x 4)T x (4 x 3)
print(hidden_layer_input, hidden_layer_input.shape)
```

```python
#h1
hidden_layer_activation = sigmoid(hidden_layer_input)
print(hidden_layer_activation)
```

```python
#z2
output_layer_input=np.dot(w_ho.T,hidden_layer_activation)
#O
output = sigmoid(output_layer_input)
```

```python
print(output)
```

```python
#Implemeanting backward propagation:

#Calculating error..tkainf sqaure error here
error=np.square(y-output)/2
print(error)
```

```python
#rate of change of error wrt output
error_wrt_output= -(Y-output)
print(error_wrt_output,"\n")

#rate of change of output wrt z2
output_wrt_z2 = np.multiply(output,(1-output))
print(output_wrt_z2,"\n")

#z1= (W_ih.T).X --> h1 = sigmoid(z1) --> z2 = (w_ho.T).h1 --> O = sigmoid(z2) --> E = (Y-O)^2/2

#rate of change of z2 wrt weights between hidden and output layer
z2_wrt_who=hidden_layer_activation
print(hidden_layer_activation)
```

```python
#Checking the shape of partial derivatives
error_wrt_output.shape, output_wrt_z2.shape,z2_wrt_who.shape
```

```python
#Shape of weights of output layer
w_ho.shape
```

```python
#rate of change of error wrt weight between hidden and output layer
error_wrt_who = np.dot(z2_wrt_who,(error_wrt_output*output_wrt_z2).T)
print(error_wrt_who,error_wrt_who.shape)
```

```python
#Rate of chnage of error wrt weight between input and hidden layer
#Error wrt w_ih
#a. rate of change of error wrt output
#b. rate of change of output wrt z2
#c. rate of change of z2 wrt hidden layer ativation (h1)
#d. rate of change of hidden layer activation wrt z1
#e. rate of change of z1 wrt weights between input and hidden layer


```

```python
#a. rate of change of error wrt output
error_wrt_output= -(Y-output)
print("error_wrt_output",error_wrt_output,"\n")

#b. rate of change of output wrt z2
output_wrt_z2 = np.multiply(output,(1-output))
print("output_wrt_z2",output_wrt_z2,"\n")

#z1= (W_ih.T).X --> h1 = sigmoid(z1) --> z2 = (w_ho.T).h1 --> O = sigmoid(z2) --> E = (Y-O)^2/2

#c. Instead of z2_wrt_who (which is h1), we are doing z2 wrt h1(hidden layer ativation)
#rate of change of z2 wrt h1
z2_wrt_h1= w_ho
print("z2_wrt_h1",z2_wrt_h1,"\n")


#d. rate of change of h1 wrt z1
h1_wrt_z1= np.multiply(hidden_layer_activation,(1-hidden_layer_activation))
print("h1_wrt_z1",h1_wrt_z1,"\n")

#e. rate of change of z1 wrt weights between input and hidden layer
z1_wrt_wih = X
print("z1_wrt_wih",z1_wrt_wih,"\n")

#z1= (W_ih.T).X --> h1 = sigmoid(z1) --> z2 = (w_ho.T).h1 --> O = sigmoid(z2) --> E = (Y-O)^2/2
#dE/dW_ih = dE/dO . dO/dz2 . dz2/dh1 . dh1/dz1 . dz1/dw_ih
print("error_wrt_output",error_wrt_output.shape,"\n")
print("output_wrt_z2",output_wrt_z2.shape,"\n")
print("z2_wrt_h1",z2_wrt_h1.shape,"\n")
print("h1_wrt_z1",h1_wrt_z1.shape,"\n")
print("z1_wrt_wih",z1_wrt_wih.shape,"\n")
```

---------------------------------------------------------------
z1= (W_ih.T).X --> h1 = sigmoid(z1) --> z2 = (w_ho.T).h1 --> O = sigmoid(z2) --> E = (Y-O)^2/2

dE/dW_ih = dE/dO . dO/dz2 . dz2/dh1 . dh1/dz1 . dz1/dw_ih

dE/dw_ih = dz1/dw_ih x [dh1/dz1 * dz2/dh1 x [dE/dO * dO/dz2]].T

        = (4 x 3) x [(3 x 3) * (3 x 1) x [(1 x 3) * (1 x 3)]].T
        = (4 x 3) x [(3 x 3) * (3 x 1) x (1 x 3)].T
        = (4 x 3) x [(3 x 3) * (3 x 3)].T
        = (4 x 3) x [(3 x 3)].T
        = (4 x 3)

        ###dE/dw_ih SHOULD BE OF SAME SHAPE AS w_ih


np.multiply is for element wise matrix multiplication
np.dot is for matrix multiplication
* operator can be used as a shorthand for np.multiply
---------------------------------------------------------------

```python
d = np.multiply (error_wrt_output,output_wrt_z2)
print(d,d.shape)
# (a x b) * (a x b) ---> result shape: (a x b)
```

```python
z=np.dot(output_wrt_z2,z2_wrt_h1) #1 x 3 & 3 x 1 --> 1
print(z)
g=np.multiply(output_wrt_z2,z2_wrt_h1) #1 x 3 multiply 3 x 1 --> 3 x 3
print(g)
```

```python
#Rate of change of error wrt weight between input and hidden layer
error_wrt_wih = np.dot(z1_wrt_wih, (h1_wrt_z1 *
  np.dot(z2_wrt_h1,(output_wrt_z2 * error_wrt_output))).T)
print(error_wrt_wih,'\n',error_wrt_wih.shape)
```

```python
#Update the parameters using gradient descent
#w = w - (alpha * dE(w)/dw)
#a. Define learning rate
lr = 0.01
print('w_ho \n',w_ho)
#b. Updating weight for output layer
w_ho = w_ho - lr * error_wrt_who
print("w_ho updated \n", w_ho)
#c. Updating weight for hidden layer
print("w_ih \n", w_ih)
w_ih = w_ih - lr * error_wrt_wih
print("w_ih updated \n", w_ih)
```

```python
#Training the model for n epochs

#Steps
#Define model architechture
#Intialize weights
#defining learning rate

epochs = 1000
```

```python
error_epoch = []
for i in range(epochs):
  #Forward propagation steps as follows
  # Hidden layer activations
  hidden_layer_input = np.dot(w_ih.T,X)
  hidden_layer_activation = sigmoid(hidden_layer_input)
  #Calculating the output
  output_layer_input = np.dot(w_ho.T, hidden_layer_activation)
  output = sigmoid(output_layer_input)

  #Backward propagation

  #Calculating error
  error = np.square(Y-output)/2
  error_wrt_output= -(Y-output)
  output_wrt_z2 = np.multiply(output,(1-output))
  z2_wrt_who = hidden_layer_activation
  ########################
  error_wrt_who = np.dot(z2_wrt_who,(error_wrt_output*output_wrt_z2).T)
  ########################
  z2_wrt_h1 = w_ho
  h1_wrt_z1= np.multiply(hidden_layer_activation,(1-hidden_layer_activation))
  z1_wrt_wih = X
  ########################
  error_wrt_wih = np.dot(z1_wrt_wih, (h1_wrt_z1 * np.dot(z2_wrt_h1,(output_wrt_z2 * error_wrt_output))).T)
  ########################

  #Updating the weights
  w_ho = w_ho - lr * error_wrt_who
  w_ih = w_ih - lr * error_wrt_wih

  # Appneding error of each epoch
  error_epoch.append(np.average(error))
```

```python
#Visualizing error after each epoch

plt.plot(np.arange(1, epochs+1),np.array(error_epoch))
```
